```python
# 1. IMPORT LIBRARIES AND SET DISPLAY OPTIONS

import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1000)
pd.set_option("display.max_rows", 100)

print("Libraries imported successfully")
```

```python
# 2. DATASET STRUCTURE ANALYSIS
data = pd.read_csv("../data/top100_ai_tools_2026.csv")

print("Shape of Dataset:", data.shape)

print("\nColumn Names:")
for col in data.columns:
    print(col)

print("\nData Types:")
print(data.dtypes)

print("\nMemory Usage:")
print(data.memory_usage())
```

```python
# 3. DUPLICATE AND DATA CLEANING CHECK

duplicate_count = data.duplicated().sum()

print("Number of Duplicate Rows:", duplicate_count)

if duplicate_count > 0:
    data = data.drop_duplicates()
    print("Duplicates removed")
else:
    print("No duplicates found")

print("Updated Shape:", data.shape)
```

```python
# 4. CATEGORY ANALYSIS

category_counts = data['Primary_Category'].value_counts()

print("Category Distribution:")
print(category_counts)

category_percent = (category_counts / len(data)) * 100

print("\nCategory Percentage Distribution:")
print(category_percent)

top_categories = category_counts.head(5)

print("\nTop 5 Categories:")
print(top_categories)
```

```python
# 5. TRAFFIC AND POPULARITY ANALYSIS

top_traffic = data.sort_values(by='Monthly_Traffic_Est', ascending=False)

print("Top 10 Tools by Traffic:")
print(top_traffic[['Tool_Name', 'Monthly_Traffic_Est']].head(10))

bottom_traffic = data.sort_values(by='Monthly_Traffic_Est', ascending=True)

print("\nBottom 5 Tools by Traffic:")
print(bottom_traffic[['Tool_Name', 'Monthly_Traffic_Est']].head(5))

traffic_stats = data['Monthly_Traffic_Est'].describe()

print("\nTraffic Statistics:")
print(traffic_stats)
```

```python
# 6. USER RATING ANALYSIS

rating_stats = data['User_Rating'].describe()

print("Rating Statistics:")
print(rating_stats)

high_rated = data[data['User_Rating'] >= 4.5]

print("\nHigh Rated Tools Count:", len(high_rated))
print(high_rated[['Tool_Name', 'User_Rating']].head())

low_rated = data[data['User_Rating'] < 3.5]

print("\nLow Rated Tools Count:", len(low_rated))
```

```python
# 7. FEATURE ENGINEERING

data['Traffic_per_User'] = data['Monthly_Traffic_Est'] / data['Active_Users_Est']

print("New Feature Created: Traffic_per_User")

print(data[['Tool_Name', 'Traffic_per_User']].head())
```

```python
# 8. RELEASE YEAR TREND ANALYSIS

year_counts = data['Release_Year'].value_counts().sort_index()

print("Tools Released Each Year:")
print(year_counts)

latest_year = data['Release_Year'].max()
earliest_year = data['Release_Year'].min()

print("\nEarliest Release Year:", earliest_year)
print("Latest Release Year:", latest_year)
```

```python
# 9. API AVAILABILITY IMPACT ANALYSIS

api_group = data.groupby('API_Available').agg({
    'Monthly_Traffic_Est': 'mean',
    'Active_Users_Est': 'mean',
    'User_Rating': 'mean'
})

print("API Availability Impact:")
print(api_group)
```
