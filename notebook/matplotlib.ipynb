```python
# 1. IMPORT LIBRARIES

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("../data/top100_ai_tools_2026.csv")

print("Dataset loaded successfully")
```

```python
# 2. PRIMARY CATEGORY DISTRIBUTION (BAR CHART)

category_counts = data['Primary_Category'].value_counts()

plt.figure(figsize=(10,6))
plt.bar(category_counts.index, category_counts.values,color='green',edgecolor='black')

plt.xlabel("Primary Category")
plt.ylabel("Number of Tools")
plt.title("Distribution of AI Tools by Primary Category")
plt.xticks(category_counts.index, rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 3. PRICING MODEL DISTRIBUTION (PIE CHART)

pricing_counts = data['Pricing_Model'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(
    pricing_counts.values,
    labels=pricing_counts.index,
    autopct='%1.1f%%'
)

plt.title("Pricing Model Distribution")
plt.show()
```

```python
# 4. RELEASE YEAR TREND (LINE PLOT)

year_counts = data['Release_Year'].value_counts().sort_index()

plt.figure(figsize=(10,6))
plt.plot(
    year_counts.index,
    year_counts.values,
    marker='o',color='black',markersize=8
)

plt.xlabel("Release Year")
plt.ylabel("Number of Tools")
plt.title("AI Tools Released Over Years")

plt.show()
```

```python
# 5. TRAFFIC VS USER RATING (SCATTER PLOT)

traffic = data['Monthly_Traffic_Est']
rating = data['User_Rating']

plt.figure(figsize=(8,6))
plt.scatter(
    traffic,
    rating,color="red",edgecolors='black'
)

plt.xlabel("Monthly Traffic Estimate")
plt.ylabel("User Rating")
plt.title("Traffic vs User Rating")

plt.show()
```

```python
# 6. ACTIVE USERS VS TRAFFIC (SCATTER PLOT)

users = data['Active_Users_Est']

plt.figure(figsize=(8,6))
plt.scatter(
    users,
    traffic,color='green',edgecolors='black'
)

plt.xlabel("Active Users Estimate")
plt.ylabel("Monthly Traffic Estimate")
plt.title("Active Users vs Monthly Traffic")

plt.show()
```

```python
# 7. STARTING PRICE DISTRIBUTION (HISTOGRAM)

prices = data['Starting_Price_USD'].dropna()

plt.figure(figsize=(8,6))
plt.hist(
    prices,
    bins=15,color='orange',edgecolor='black'
)

plt.xlabel("Starting Price (USD)")
plt.ylabel("Frequency")
plt.title("Distribution of Starting Prices")

plt.show()
```

```python
# 8. CATEGORY VS AVERAGE RATING (BAR CHART)

avg_rating_category = data.groupby('Primary_Category')['User_Rating'].mean()

plt.figure(figsize=(10,6))
plt.bar(
    avg_rating_category.index,
    avg_rating_category.values,edgecolor='black',color='darkred'
)

plt.xlabel("Primary Category")
plt.ylabel("Average User Rating")
plt.title("Average Rating per Category")
plt.xticks(avg_rating_category.index, rotation=45)

plt.show()
```
