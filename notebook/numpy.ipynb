```python
import pandas as pd
import numpy as np

# Load dataset
data = pd.read_csv("../data/top100_ai_tools_2026.csv")

# Select numeric columns
numeric_data = data.select_dtypes(include=np.number)

print("Numeric Columns:")
print(numeric_data.columns)
```

```python
# Basic Statistics
print("\nMean:")
print(np.mean(numeric_data, axis=0))

print("\nMedian:")
print(np.median(numeric_data, axis=0))

print("\nVariance:")
print(np.var(numeric_data, axis=0))

print("\nStandard Deviation:")
print(np.std(numeric_data, axis=0))

print("\nMinimum Values:")
print(np.min(numeric_data, axis=0))

print("\nMaximum Values:")
print(np.max(numeric_data, axis=0))
```

```python
# correlation matrix
corr_matrix = np.corrcoef(numeric_data.T)

print("\nCorrelation Matrix:")
print(corr_matrix)
```

```python
# ranking tools by traffic
traffic = numeric_data['Monthly_Traffic_Est'].values

sorted_indices = np.argsort(traffic)[::-1]

print("\nTop 5 Tools by Traffic:")
print(data.iloc[sorted_indices[:5]][['Tool_Name', 'Monthly_Traffic_Est']])
```

```python
# Skewness Analysis
traffic = numeric_data['Monthly_Traffic_Est'].values

mean = np.mean(traffic)
median = np.median(traffic)

print("\nTraffic Mean:", mean)
print("Traffic Median:", median)

if mean > median:
    print("Traffic distribution is right-skewed")
else:
    print("Traffic distribution is left-skewed")
```

```python
# normalize the traffic data
min_vals = np.min(numeric_data, axis=0)
max_vals = np.max(numeric_data, axis=0)

normalized = (numeric_data - min_vals) / (max_vals - min_vals)

print("\nNormalized Data (First 5 Rows):")
print(normalized[:5])
```

```python
#percentiles
print("\n25th Percentile:")
print(np.percentile(numeric_data, 25, axis=0))

print("\n50th Percentile (Median):")
print(np.percentile(numeric_data, 50, axis=0))

print("\n75th Percentile:")
print(np.percentile(numeric_data, 75, axis=0))
```
